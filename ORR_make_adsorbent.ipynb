```python
from ase.visualize.plot import plot_atoms
import ase.io
from ase import Atom, Atoms

#   
slab = ase.io.read("./POSCAR") 

plot_atoms(slab)
```

## OOH 

```python
site1 = slab.copy()
oxygen_position = slab[18].position+ [0,0,1.5] # site atom    
#numbering 0  !

new_atom = Atom('O', position = oxygen_position)
site1.append(new_atom)
new_atom_2 = Atom ('O', position = new_atom.position + [0,0,1.4])
site1.append(new_atom_2)
new_atom_3 = Atom ('H', position = new_atom_2.position + [0,0,1])
site1.append(new_atom_3)
plot_atoms(site1, rotation = ("-90x"))
#plot_atoms(site1, )
```

###  

```python
site1.write ("./3_OOH_C/INPUT/POSCAR", vasp5 = True, direct = True)
```
